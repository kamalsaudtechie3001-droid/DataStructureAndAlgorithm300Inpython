We can achieve a more optimal solution by recognizing runs of zeroes as we iterate over the array. For each segment of contiguous zeroes, the number of zero-filled subarrays can be calculated using combinatorial arithmetic.

We can find out the number of subarray filled zero by using the formula of sum of n natural number.
 n(n+1) //2 = 1 + 2 + 3 + ...... + .. n

```python
def zeroFilledSubArray(nums):
    n = len(nums)
    output = 0
    numberOfZero = 0
    #To iterate the array to find eithe number id=s zero or not.
    for num in nums:
        if num == 0:
            numberOfZero += 1
        else:
            output += numberOfZero *( numberOfZero + 1)//2
            numberOfZero = 0
    
    output += numberOfZero *( numberOfZero + 1)//2  
    return output

nums = [0,0,0,2,0,0]
answer = zeroFilledSubArray(nums)
print(f"The number of subarray filled with zero are:" , answer)
```
