**Better solution to move zero using two pass approach**


```python
def moveZeroTwoPass(nums):
    n = len(nums)
    rightPointer = 0
    # First pass
    for i in range(n):
        if nums[i] != 0:
            nums[rightPointer] = nums[i]
            rightPointer += 1
    # second pass
    while rightPointer < n:
        nums[rightPointer] = 0
        rightPointer += 1
    
    return nums

# The given array where we are moving all the non- zero elements to left of the array.
nums = [0 , 1, 0,3,12]
print(f" The original array is:" , nums)
ans = moveZeroTwoPass(nums)
print(f"The resultant array after move zero to right side is: " , ans)
```
