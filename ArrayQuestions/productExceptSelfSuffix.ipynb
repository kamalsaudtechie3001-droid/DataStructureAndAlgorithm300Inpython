In this approach, we are creating three array of size equal to the size of the array.i.e prefix , suffix and result array.

```python
def productExceptSelf(nums):
    n = len(nums)
    output = [0] * n
    suffix = [0] * n
    prefix = [0] * n

    # To calculate product of the array from the left to right of the array.
    prefix[0] = 1 
    for i in range(1 , n):
        prefix[i] = prefix[i-1] * nums[i-1]

    #To calculate the product of the element from right to left of the array.
    suffix[n-1] = 1
    for  j in range(n-2 , -1 , -1):
        suffix[j] = suffix[j + 1] * nums[j + 1]
    
    #To calculate the product of the array except self
    for i in range(n):
        output[i]  = prefix[i] * suffix[i]
    return output


nums  = [-1,1,0,-3,3]
ans = productExceptSelf(nums)
print(f"The product of the array except self is:" , ans)
```
