```python
def rotateArrayExtraArray(nums , k):
    n = len(nums)
    result = [0] * n
    k = k%n
    for i in range(n):
        result[(i + k) % n] = nums[i] # The index belongs to 0 to n -1 index. if value exceeds than n -1.
    

    #Copy the value of result array to the original array.
    for i in range(n):
        nums[i]  = result[i]
    
    return nums

nums = [1, 2, 3, 4,5,6 ,7]
print(f"The array before rotation is:", nums)
k = 3
ans = rotateArrayExtraArray(nums , k)
print(f"The rotated array is:" , ans)
```

**Time Complexity** : O(N) 

There is a single loop to traverse the array

**Space Complexity** : O(N)

Because We are using result array of length N to store the elements.
