In this approach we are creating a set. then all elements of given array to the set except the duplicate elements then check which  positive number is missing from 1 to n +2 (array of length + 2).

```python
def firstMissingElement(nums):
    n = len(nums)
    set_nums = set() # Create a Set to store unique elements of the array.
    for num in nums:
        set_nums.add(num)
    
    # To check the either number is present or not 
    for i in range(1 , n + 2):
        if i not in set_nums:
            return i
    return -1


nums = [7,8,9,11,12]
ans = firstMissingElement(nums)
print(f"The first missing positive number is:" , ans)
```

**Time Comlexity** : O(N) + O(N) = O(N) There is not any overlaping loop in the program.

**Space Complexity** : O(N) We are using data structure set to store the all the unique elements of the array. in worst case scenarion. We are using the N Size set to store all the elements.
