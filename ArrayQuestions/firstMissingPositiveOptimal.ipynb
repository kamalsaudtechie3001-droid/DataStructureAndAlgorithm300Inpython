Given an unsorted integer array nums. Return the smallest positive integer that is not present in nums.

You must implement an algorithm that runs in O(n) time and uses O(1) auxiliary space.

Example 1:

Input: nums = [1,2,0]

Output: 3

Explanation: The numbers in the range [1,2] are all in the array.

```python
def firstMissingPositive( nums):
        n = len(nums)
        #To place right elements in right place index 0 contains the element 1 and 1 have 2 so on..
        for i in range(n):
            while 1 <= nums[i] <= n and nums[nums[i] - 1] != nums[i]:
                target = nums[i] - 1
                nums[target] , nums[i]  = nums[i] , nums[target]
        

        #To check either elements is present in right place.
        for i in range(n):
            if nums[i] != i +1:
                return i + 1
        
        return n + 1

nums = [3,4,-1,1]
ans = firstMissingPositive(nums)
print(f"The missing positive number is:" , ans)
```
