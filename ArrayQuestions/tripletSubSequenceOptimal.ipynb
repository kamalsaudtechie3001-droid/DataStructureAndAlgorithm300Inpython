As we scan through the array, the goal is to gradually refine these candidates. If at any point we encounter a number greater than both first and second, it becomes our third, and the triplet exists.

```python
def tripletSubSequenceOptimal(nums):
    first = float("inf")
    second = float("inf")
    for num in nums:
        if num <= first:
            first = num
        elif num <= second:
            second  = num
        else:
            return True
        


    return False

nums = [5,4,3,2,1]
ans = tripletSubSequenceOptimal(nums)
if ans:
    print("Triplet sum exits:")
else:
    print("Triplet sum does not exits:")
```

**Time Complexity** : O(N)

Iterating the array elements through single loop.
