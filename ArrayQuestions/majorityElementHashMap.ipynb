**HashMap**

Hashmap is use to store data in the form of keys and values pairs. Key is the unique but value is the number of element present in the array.

```python
def majorityElementHashMap(nums):
    n = len(nums)
    freq = {} # Create a dictionery to store key and value
    # To check the frequency of the number are in the array
    for num in nums:
        if num in freq: # Either num is in the hashmap or not. if present increse frequency of the number by one not make it one.
            freq[num] += 1
        else:
            freq[num] = 1
        #To check the frquency of number greater than the half of its length.
        if freq[num] > n//2:
            return num
    return -1

nums = [2, 2, 1 ,1 , 1 ,2 ,2]
ans = majorityElementHashMap(nums)
print(f"The majority element is:" , ans)
```

**Time Complexity** : O(n)

The time complexity of the code, I  have written is big of N this is because we are iterating through single for loop in the worst case scenario it will be big of n.

**Space Complexity** : O(n)

This is because we are using another data structure to store keys and values. Which is hashmap or dictionery.
