Given an integer array nums, return an array answer such that answer[i] is equal to the product of all the elements of nums except nums[i].

```python
def productExceptSelf(nums):
    #Calcilate length of the array.
    n = len(nums)
    #To array to store all the product element
    output = [0] * n
    #Product of the array at index 0 to n-1.

    for i in range(n):
        output[i] = 1
        #Inner loop to to traverse array to calculate product of all the elements in an array
        for j in range(n):
            if i != j:
                output[i] *= nums[j]
    
    #To store all the elements from output array to original array.
    for i in range(n):
        nums[i] = output[i]
    return nums

nums = [1,2,3,4]
result = productExceptSelf(nums)
print(f"The resultant array after product of all elements except itself is:" , result)
```

**Time Complexity** :  O(N)

**Space Complexity** : O(1) There is no extra space used to run the program.
