Given an unsorted integer array nums. Return the smallest positive integer that is not present in nums.

You must implement an algorithm that runs in O(n) time and uses O(1) auxiliary space.

```python
#Brute Force Approach.
def firstMissingPositive(nums):
    n = len(nums)
    #To iterating the array.
    for i in range(1 , n +2):
        found = False
        for num in nums:
            if num == i:
                found = True
                break
    
        if not found:
          return i
    return -1

nums = [3,4,-1,1]
ans = firstMissingPositive(nums)
print(f"The first missing positive is:", ans)
```

**Time Complexity** : O(N2)

Two loop are using in this program but one loop nested inside the other loop.

**Space Complexity** : O(1) 

There is no extra memory is used. So, space complexity is constant.
