Given an integer array nums , Move all the zero to right side of the array. while maintaining the relative order of the non zero elements.

```python
def movezeroBruteForce(nums):
    n =len(nums) # length of the array.
    result  = [0] * n # Create a resultant array initialize with the element zero
    j = 0 # Pointer to  track the non-zero elements.
    for i in range(n): # To store all the non zero elements to the result array.
        if nums[i] != 0:
            result[j] =  nums[i]
            j += 1
    
    #To store all the zero elements in the remaining places.
    for k in range(j , n):
        result[k] = 0
    
    # To bring back all the elements from result array to original array.
    for i in range(n): #By default start from zero.
        nums[i] = result[i]
    
    return nums


# The given array where we are moving all the non- zero elements to left of the array.
nums = [0 , 1, 0,3,12]
print(f" The original array is:" , nums)
ans = movezeroBruteForce(nums)
print(f"The resultant array after move zero to right side is: " , ans)


```
