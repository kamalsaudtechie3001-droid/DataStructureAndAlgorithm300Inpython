**Peak- Valley approach to get maximum profit while buy and sell of stock in multiple times**

```python
# Peak Valley Approach.
def bestBuySell(prices):
    n = len(prices)
    peak = prices[0]
    valley = prices[0]
    maxProfit = 0
    i = 0
    while i < n - 1:
        # To calculate the minimum value at valley.
        while i < n-1 and prices[i] <= prices[i + 1]:
            valley = prices[i]
            i += 1

        # To calculate the maximum value at peak.
        while i < n-1 and prices[i] >= prices[i+1]:
            peak = prices[i]
            i +=1
        maxProfit  += peak - valley

    return maxProfit

prices = [7,1,5,3,6,4] 
result = bestBuySell(prices)
print(f" The maximum profit again to buy and sell of the stock:" , result)
```

**Time Complexity** : O(N)

Iterating the price array once.

**Space Complexity** : O(1) There is constant space is used.
