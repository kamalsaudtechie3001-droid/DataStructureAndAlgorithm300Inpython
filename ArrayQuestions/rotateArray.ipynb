In this problem, Given an array nums of length N and key value k means we need to  right rotate of an array by k position. While k is the non-negative integer.

```python
#Bruteforce Approach 
# In this approach we are going to keep the last element in an variable and  shift remaining elements right to the array and,
# put  back last element we keep on the variable at 0's index in each iteration.

def rotateArray(nums , k):
    n = len(nums)
    k = k % n # To handle the situation, when the value of k is greater than or equal to n.
    #This shows the number of iteration to shift operation is used.
    for i in range(k):
        #This is the variable where we are holding the last element of the array.
        previous = nums[n-1]
        #Loops to shift all the remaining element to right of the array.
        for j in range(n-1 , 0 , -1):
            nums[j] = nums[j-1]
        
        nums[0] = previous
    
    return nums

nums = [1, 2, 3, 4,5,6 ,7]
print(f"The array before rotation is:", nums)
k = 3
ans = rotateArray(nums , k)
print(f"The rotated array is:" , ans)

```

**Time Complexity** : O(K *N) 

This is because there are two overlaping loop one of length k and another of N.

**Space Complexity** : O(1)

There is constatnt space complexity. No extra space is used to run the program.
