Given an integer array nums, return the number of subarrays filled with 0.

```python
#Brute Force Approach to how many zero filled contiguous sub array we should make.

def zeroFilledSubArray(nums):
    n = len(nums)
    count = 0
    for i in range(n):
        j = i
        while j < n and nums[j] == 0:
            count += 1
            j += 1
    
    return count

nums = [1,3,0,0,2,0,0,4] 
ans = zeroFilledSubArray(nums)
print(f"The number of zero filled sub array are :" , ans)
```

**Time Complexity** : O(N2)

There is the overlaping loop. So,the time complexity of the is quadratic.

**Space Complexity** : O(1) 

There is no extra space is uses to run this code.
