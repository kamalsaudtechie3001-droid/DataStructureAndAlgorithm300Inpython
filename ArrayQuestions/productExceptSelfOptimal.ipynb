**Optimal approach**:


```python
#To Calculate the product of the array except the self.

def productElements(nums):
    n = len(nums)
    output = [0] * n
    output[0] = 1
    for i in range(1 ,n):
        output[i] = output[i-1] * nums[i - 1]
    right = 1
    for i in range(n):
        output *= right
        right *= nums[i]
    
    return output

nums = [1,2,3,4]
ans = productElements(nums)
print(f"The product of the elements is:", ans)
```
