The main idea behind this approach is that we are trying to find minimum price of the stock in the array. This shows that minimal the price to buy and maximum price to sell to get maximum profit. 


```python
# Optimal Approach.
def bestTimeToBuyAndSellStock(prices):
    n = len(prices)
    minPrice = float("inf") # Maximum the minPrice initially to minimize the value.
    maxProfit = 0
    #To Traverse the prices array to get maximum profit.
    for i  in range(n):
        if minPrice > prices[i]:
            minPrice = prices[i]
        profit = prices[i] - minPrice
        if profit > maxProfit:
            maxProfit = profit
    return maxProfit

prices = [7,1,5,3,6,4]
maximumProfit = bestTimeToBuyAndSellStock(prices)
print(f"The maximum profit to buy and sell of the stock is:", maximumProfit)
```

**Time Complexity** : O(N)

**Space Complexity** : O(1) 

There is constant time complexity. means no extra memory is used.
