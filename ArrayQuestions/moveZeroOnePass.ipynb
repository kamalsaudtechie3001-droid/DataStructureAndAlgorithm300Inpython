**One pass Solution of the move Zero problem**

```python
def moveZeroOnePass(nums):
    n = len(nums)
    onepointer = 0
    # To scan the array.
    for secondpointer in range(n):
        if nums[secondpointer] != 0:
            if onepointer != secondpointer:
                # Swap of array elements.
                nums[onepointer],nums[secondpointer]  = nums[secondpointer],nums[onepointer]
            onepointer += 1
    
    return nums


# The given array where we are moving all the non- zero elements to left of the array.
nums = [0 , 1, 0,3,12]
print(f" The original array is:" , nums)
result = moveZeroOnePass(nums)
print(f"The resultant array after move zero to right side is: ", result)
```
