In this problem given an array  nums and size is n. To print the number of unique elements presents in this array . With linear time and space complexity. I am going to solve this problem using two pointer approach.

```python
def removeDuplicateElement(nums):
    n = len(nums)
    #pointer1 is the index, which is used to store the unique elements in the array.
    pointer1 = 0
    #To trace the arrray by using second pointer.
    for pointer2 in range(1 ,n):
        if nums[pointer1] != nums[pointer2]:
            pointer1 += 1
            nums[pointer1] = nums[pointer2]
    

    return pointer1 + 1

nums = [0 ,0, 1, 1, 1 ,2, 2, 3, 3]
ans = removeDuplicateElement(nums)
print(f"The number of  unique element is in the array after remon=ving the duplicate. ", ans)
```

**Time Comlexity** : O(N) There is no any overlaping loop.

**Space Complexity** : O(1) . While running this program we are not using any extra memory. So, space complexity to be linear.
