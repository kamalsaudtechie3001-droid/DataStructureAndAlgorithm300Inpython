```python
# Reverse arry functions
def reverse( nums , start , end):
    while start < end:
        nums[start] , nums[end] = nums[end] , nums[start]
        start += 1
        end -= 1


def rotateArrayOptimal( nums , k):
    n = len(nums)
    k = k %n 
    # To reverse all the array.
    reverse(nums , 0 , n-1)
    #To reverse thearray of length k
    reverse(nums , 0 , k-1)
    #To reverse the remaining array.
    reverse(nums , k , n-1)

    return nums


nums = [1, 2, 3, 4,5,6 ,7]
print(f"The array before rotation is:", nums)
k = 3
ans = rotateArrayOptimal(nums , k)
print(f"The rotated array is:" , ans)


    
```

**Time Complexity** : O(N)

**Space Complexity** : O(1)
