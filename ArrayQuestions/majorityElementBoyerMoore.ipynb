**Boyer Moore Voting Algorithm**

The algorithm works by "canceling out" pairs of different elements. Since the majority element appears more than half the time, even if every other element in the array pairs up with a majority element to cancel it out, at least one majority element will remain at the end. 

```python
def majorityElementBoyerMoore(nums):
    n = len(nums)
    candidate = nums[0]
    count = 0
    #Traverse the array 
    for num in nums:
        if count == 0:
            candidate = num
        #This is the process of cancellation of equally appear element.
        if num == candidate:
            count += 1
        else:
            count -=1
    
    return candidate #After the all the cancellation of equal appearence in the array remaining element is the majority element.

nums = [ 2, 2, 1, 1, 1 , 2, 2]
ans = majorityElementBoyerMoore(nums)
print(f"The majority element is :" , ans)
```

**Time Complexity** : O(N) 

There is only loop to traverse the array. If the size of the array is n at that situation the time complexity is big of N.

**Space Complexity** : O(1)

There is no any extra memory is used. So, space complexity is linear.
