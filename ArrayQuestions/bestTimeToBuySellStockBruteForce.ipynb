There is given  an array nums with size n , in this array our task is to calculate the best time to buy stock, after that  we also need to find appropriate time to sell it to gain maximum profit on it. First we buy than sell. We are traversing the array to buy stock first than sell stock in the array of price of each stock is given to us.

```python
#Brute Force Approach
def bestTimeBuyAndSell(nums):
    n = len(nums)
    maxProfit = 0
    #To traverse the array to get the buying price.
    for i in range(n):
        # This is the array traverse to find the selling price of the stock.
        for j in range(i + 1 , n):
            profit = nums[j] - nums[i]
            if profit > maxProfit:
                maxProfit = profit
    
    return maxProfit

prices = [7,1,5,3,6,4]
maxProfit = bestTimeBuyAndSell(prices)
print(f"The maximum profit to earn to buy and sell of the stock in the price array:" , maxProfit)
```

**Time Complexity** : O(N2) 

The complexity of the program is the big of N * N or (N2), Because there is an overlaping loop.

**Space Complexity** : O(1)

 There is noo extra space is being used while running the program.
