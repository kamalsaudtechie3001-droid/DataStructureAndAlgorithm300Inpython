Given an array nums of size n and return majority element in it.

The majority element is that the elements that appears more than [N/2] times.

```python
def majorityElementBruteForce(nums):
    # Size of the array or nums
    n = len(nums)
    # First loop that iterate each elements.
    for i in range(n):
        count = 1
        # Second loop to find the elements.
        for j in range( i + 1 , n):
            if nums[i] == nums[j]:
                count += 1
        if count > n // 2:
         return nums[i]
    
    return -1

nums = [2, 2, 1 ,1 ,1, 2, 2]
ans = majorityElementBruteForce(nums)
print(f" The majority elements that present in the array is " , ans)
```

### Time complexity = o(n2) 
Because there is an overlapping loop are is in the code.

 ### Space Complexity = o(1)

There is no extra memmory is used.
