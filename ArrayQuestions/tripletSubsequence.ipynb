Given an integer array nums, return true if there exists a triple of indices (i, j, k) such that i < j < k and nums[i] < nums[j] < nums[k]. If no such indices exists, return false.

```python
def tripletSubSequence(nums):
    n = len(nums)
    #Traverse the loop to iterate first element
    for i in range(n-2):
        #To itrate the loop to get second number.
        for j in range( i+1 , n-1):
            #Third number 
            for k in range(j+1 ,n):
                if nums[i] < nums[j] < nums[k]:
                    return True
    return False


nums = [2,1,5,0,4,6]
ans = tripletSubSequence(nums)
if ans:
    print("The triplet subsequence is present")
else:
    print("The triplet subsequence is not present")
```

**Time Complexity** : O(N3)

There are three loops are being overlap to one another.

**Space Complexity** : O(1) We are not using any extra memory in this program.
