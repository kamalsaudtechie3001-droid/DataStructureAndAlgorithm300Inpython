```python
def maxProfitBuyAndSellStock(nums):
    n = len(nums)
    maxProfit = 0
    minPrice = float("inf")

    for i in range(n):
        if nums[i] < minPrice:
            minPrice = nums[i]
        profit = nums[i] - minPrice
        if profit > maxProfit:
            maxProfit = profit
    
    return maxProfit

prices = [7,1,5,3,6,4]
ans = maxProfitBuyAndSellStock(prices)
print("The maximum profit is :" , ans)



    
```

**Time Complexity** : O(N) where n is the length of the list.

**Space Complexity** : O(1) There is no use of extra memory.
