You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists into one sorted list. The list should be made by splicing together the nodes of the first two lists.

Example :

Input: list1 = [1,2,4], list2 = [1,3,4]

Output: [1,1,2,3,4,4]







**Head**

- The head is the first node of the linked list

- Its the entry point to the list

- If the list is empty  head = None

**Tail**

- The tail is the last node of the linked list

- Its next pointer is None

- Keeping a tail makes insertion at the end O(1) instead of O(n)

- List stores value only and linked list store both value and reference pointer to point to next value

```python
#Define listnode before working on the linked list
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def mergeLinkedList(list1 , list2):
    #Create a dummy 
    dummy = ListNode()
    tail = dummy
    # If the list1 and list two have some value

    while list1 and list2:
        if list1.val <= list2.val:
            tail.next = list1
            list1 = list1.next
        else:
            tail.next = list2
            list2 = list2.next
        tail = tail.next 
    
    # Add remaining node of the non-empty list
    if list1: #  iF first list is not empty
        tail.next = list1

    if list2: # If second list is not empty
        tail.next = list2

    return dummy.next # return the merge list
   
def buildLinkedList(arr):
    dummy = ListNode()
    curr = dummy
    for v in arr:
        curr.next = ListNode(v)
        curr = curr.next
    return dummy.next

list1 = buildLinkedList([1,2,4])

list2 = buildLinkedList([1,3,4])


ans = mergeLinkedList(list1 , list2)

# Functions to print in the linked list.
def printLinkedList(head):
    curr = head
    while curr:
        print(curr.val, end=" -> ")
        curr = curr.next
    print("None")

printLinkedList(ans)


    
```

**Time Complexity** : O(N + M) Where m and n are the length of first and second list.

**Space Complexity** : O(1) We are not creating extra linked list but just create dummy node only.
