# Two Sum 
Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

Input: nums = [2,7,11,15], target = 9

Output: [0,1]

Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

```python
def twoSum( nums :list[int] , target :int):
    # Create a dictinery that store value and index as well.
    prevMap ={}
    # To iterate the list to store the value and index.
    for i , num in enumerate(nums):
        diff = target - num
        # To check either number is in map or not
        if diff in prevMap:
            return [prevMap[diff] , i] # To store the current index and index at which i get the number.
        # if number is not in the map . we can store value and index.
        prevMap[num] = i
    
    return

nums = [2 , 1, 5 ,3]
target = 4
result = twoSum(nums , target)
print(f"The index of the two number that sum " , result)

```

**Time complexity** :

O(N). where n is the length of nums in the worst case we can travel whole list.

**Space Complexity** : O(N)

The linear space is used while running this program because of hashmap.

**Enumerator** : Gives us both value and index.
