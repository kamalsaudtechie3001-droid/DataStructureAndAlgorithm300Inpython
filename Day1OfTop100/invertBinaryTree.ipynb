# Invert Binary Tree....

Given the root of a binary tree, invert the tree, and return its root.



Example: 

Input: root = [4,2,7,1,3,6,9]



Output: [4,7,2,9,6,3,1]

```python
#Invert Binary tree

class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


    def invertBinaryTree(self, root):
        # To check root is empty or not
        #base case
        if not root:
            return None
        #Swap the left and right node of the binary tree
        #creat temp variable to hold the value of left node of the tree
        temp = root.left
        root.left = root.right
        root.right = temp
        # Recursion on the left sode of the binary tree.
        left = self.invertBinaryTree(root.left)
        #Recursion on the right side of binary tree
        right = self.invertBinaryTree(root.right)

        return root
    
    # Function to print tree in-order (left, root, right)
def printInOrder(root):
    if root:
        printInOrder(root.left)
        print(root.value, end=' ')
        printInOrder(root.right)

# Example usage:
if __name__ == "__main__":
    # Create a binary tree
    root = TreeNode(1)
    root.left = TreeNode(2)
    root.right = TreeNode(3)
    root.left.left = TreeNode(4)
    root.left.right = TreeNode(5)
    root.right.left = TreeNode(6)
    root.right.right = TreeNode(7)

    print("Original tree in-order:")
    printInOrder(root)
    print("\n")

    # Invert the tree
    tree = TreeNode(0)  # we just need an instance to call the method
    tree.invertBinaryTree(root)

    print("Inverted tree in-order:")
    printInOrder(root)

    


```
