Given an integer array of size n, find all elements that appear more than ⌊ n/3 ⌋ times.

```python
def majorityElement( nums):
        n = len(nums)

        
        freq ={}
        for num in nums:
            if num in freq:
                freq[num] += 1
            else:
                freq[num] = 1
        #To find the elements in an array which can appear more than n // 3 times.
        # n//3 means there is max two elements have count more than n//3. 
        result = []
        for num, count in freq.items():
            if count > n // 3:
                result.append(num)
        return result

nums = [3,2,3]
ans = majorityElement(nums)
print(f"Tha resultant array is:" , ans)
```

**Time Complexity** : O(n)

**Space Complexity**: O(N) This is because we are using dictionery to store key and values.
