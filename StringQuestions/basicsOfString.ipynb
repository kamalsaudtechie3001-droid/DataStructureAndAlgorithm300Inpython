Some of the common examples of the strings are shown below.


```python
s1 = "Hello"
s2  = "World"
s = """This is 
the example of the multi-line string"""
```

**Creating String**

```python
Name = "Kamal"
s = "" #Empty String
char = "A" #Single character of the string

#Access of the character of the string
print(Name[0]) #To Print the first character of the string.
```

**String Slicing**

```python
text = "Python"
n = len(text)
print(text[:n])
print(text[1:5])
print(text[1:3])
print(text[::-1]) #To reverse the string start from the last elements end after first element and decreasing by one character.
```

**String Immutability**

```python
text = "Python"
# text[0] = "J"  ‚ùå Error
```

```python
text = "J" + text[1:] #We could not change the string character but we can add character on it.
print(text)
```

**String Operator**

```python
a  = "Hello"
b = "World"
final = a +b
print(final)
print(a *3) #Print the value of a three times.
print("W" in b) #Print either True Or False.
```

**Common String Methods**


```python
s = "  Python is a dynamically typed programming language "
#UpperCase.
print(s.upper())
print(s.lower()) # Lower case.
print(s.strip()) # To remove the space.
print(s.lstrip()) #To remove space of left side of the string .
print(s.rstrip()) #To remove the space of the right of the string.
print(s.split())
print("" .join(s))
print(s.replace("Python" , "Jave"))
print(s.find("is")) # To print the starting index of the world.
print(s.index("P")) # To print the index of the char.
```

**String Formatting**

F-String

Format

```python
name = "Ganesh"
age = 33
print(f"The name is { name} and is {age} years old")
print("The name is {} and is {} years old".format(name, age))
```

**Checking String Content**

```python
s = "Python123"

print(s.isalpha()) #Check alphabhets
print(s.isdigit()) # check digit
print(s.isalnum())# Check alphabhets and numeric
print(s.islower())
print(s.isupper())
```

**Looping In String**

```python
for char in "Python":
    print(char)
```
