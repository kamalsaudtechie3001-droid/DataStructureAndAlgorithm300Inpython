A phrase is a palindrome if, after converting all uppercase letters into lowercase letters and removing all non-alphanumeric characters, it reads the same forward and backward. Alphanumeric characters include letters and numbers.

Given a string s, return true if it is a palindrome, or false otherwise. 

```python
def isPalindrome(s):
    left = 0 # To initialize the first string.
    right = len(s) -1
    # To traverse the entire input string
    while left < right:
        # To filter or ignoring the character other than the alphabhet and number from left to right of the string
        while left < right and not s[left].isalnum():
            left += 1
        
        # To filter or ignoring the character other than the alphabhet and number from left to right of the string
        while left < right and not s[right].isalnum():
            right -= 1
        
        # To check left and right charater in a string
        if s[left].lower() != s[right].lower(): # To convert in lowercase as well.
            return False
        left +=1 
        right -= 1
    return True

s = "A man, a plan, a canal: Panama"
result = isPalindrome(s)
if result:
    print(f"The string you entered is palindrome:")

else:
    print(f"The string you entered is not panlindrome:")
```
